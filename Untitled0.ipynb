```python
%%writefile /content/train_model.py
import torch
import os
from transformers import GPT2LMHeadModel, GPT2Tokenizer, Trainer, TrainingArguments
from datasets import Dataset

# ✅ Disable WandB (Prevents unnecessary logging)
os.environ["WANDB_DISABLED"] = "true"

# ✅ Load GPT-2 Model & Tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# ✅ Fix Padding Token Issue
tokenizer.pad_token = tokenizer.eos_token  # Set pad token to eos token

# ✅ Load Custom Dataset
def load_data(file_path="/content/dataset.txt"):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            lines = f.readlines()
        return Dataset.from_dict({"text": lines})
    except FileNotFoundError:
        print("⚠️ Dataset file not found! Ensure '/content/dataset.txt' exists.")
        return None

dataset = load_data()

# ✅ Tokenization Function (Correct Label Formatting)
def tokenize(example):
    encoding = tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=512
    )
    encoding["labels"] = encoding["input_ids"]  # ✅ Correct label formatting
    return encoding

if dataset:  # ✅ Ensure dataset exists before processing
    dataset = dataset.map(tokenize)

    # ✅ Training Arguments (Optimized)
    training_args = TrainingArguments(
        output_dir="/content/results",
        report_to="none",  # ✅ Prevent WandB logging
        num_train_epochs=3,
        per_device_train_batch_size=2,
        save_steps=500,
        save_total_limit=2,
        logging_dir="/content/logs"
    )

    # ✅ Initialize Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=dataset
    )

    # ✅ Start Fine-Tuning (with Error Handling)
    try:
        trainer.train()
        print("✅ Training Completed Successfully!")
    except Exception as e:
        print(f"⚠️ Training Error: {e}")

    # ✅ Save Fine-Tuned Model
    model.save_pretrained("/content/fine_tuned_gpt2")
    tokenizer.save_pretrained("/content/fine_tuned_gpt2")

    print("🚀 Model saved at '/content/fine_tuned_gpt2'")
```
